4. Lag et scatter-plott med antall biler som er over 20 år og prosentvis andel biler som er vraket. Ser det ut til å være en sammenheng?

```python
# Importerer numpy
import numpy as np

# Lager nye kolonner med naturlig logaritme av spesifiserte kolonner
# Legger til en liten konstant (i dette tilfellet 1) for å unngå log(0) og håndtere data som kan være 0.
df['log_Over_20_år'] = np.log(df['Over 20 år'] + 1)
df['log_Prosent_vraket'] = np.log(df['Personbiler vraket mot pant (prosent av bestanden)'] + 1)

# Setter figurdimensjonene
plt.figure(figsize=(12, 7))

# Plotter scatter-plott for de log-transformerte variablene
plt.scatter(
    df['log_Over_20_år'], 
    df['log_Prosent_vraket'], 
    color='green', 
    edgecolor='black', 
    alpha=0.6,
    s=50  # setter markørstørrelse 
)

# Setter tittel og etiketter for aksene
plt.title('Scatter-plot: Log av biler over 20 år vs Log av prosentvis andel vraket biler', fontsize=15)
plt.ylabel('Log av prosentvis andel vraket biler', fontsize=12)
plt.xlabel('Log av antall biler over 20 år', fontsize=12)

# Legger til rutenett for bedre lesbarhet
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Justerer layout for å passe alle elementene
plt.tight_layout()

# Viser diagrammet
plt.show()
```

5. Lag nye variabler med den naturlige logaritmen for antall biler som er over 20 år og prosentvis andel biler som er vraket, og lag et scatter-plott av disse. Ser det nå ut tilå være en sammenheng?

```python
# Beregner korrelasjonskoeffisienten mellom de to log-transformerte variablene
correlation_coefficient = df['log_Over_20_år'].corr(df['log_Prosent_vraket'])

# Viser scatter-plot for visuell inspeksjon
plt.figure(figsize=(12, 7))
plt.scatter(df['log_Over_20_år'], df['log_Prosent_vraket'], color='purple', edgecolor='black', alpha=0.6)
plt.title('Scatter-plot: Log-transformerte verdier', fontsize=15)
plt.ylabel('Log av prosentvis andel vraket biler', fontsize=12)
plt.xlabel('Log av antall biler over 20 år', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

print(f"Korrelasjonskoeffisienten er: {correlation_coefficient:.2f}")
if 0.7 <= correlation_coefficient <= 1:
    print("Ja, basert på den beregnede korrelasjonskoeffisienten ser det ut til å være en sterk positiv sammenheng mellom den logaritmiske transformasjonen av antall biler som er over 20 år og den logaritmiske transformasjonen av prosentvis andel vrakede biler.")
else:
    print("Basert på den beregnede korrelasjonskoeffisienten er det ikke en sterk positiv sammenheng mellom variablene.")
```

1. Lag tilbuds- og etterspørselsfunksjoner for Tesla. Prøv å benytte ikke-lineære funksjoner hvis mulig.

Formen funksjonene kan du bestemme selv, men de må være slik ati. 

En likevekt kan bestemmii. es.
Likevektsprisen er omtrent det en Tesla koster i dag.

```python
import sympy as sp

# Definerer variabler for penere kode
P = sp.symbols('P')
a, b, c, d = 60000, 0.0001, 0.0001, 0  #Dette er bare eksempelverdier og kan justeres

# Etterspørselsfunksjon
Qd = a - b*P**3

# Tilbudsfunksjon
Qs = c*P**3 - d
```

```python
2.Lag en funksjon som
i. tar en etterspørsel- og tilbudsfunksjon, som argument (altså to argumenter)
ii. som regner ut likevekten med Sympy
iii. som tegner tilbud og etterspørselskfunksjonene, med konsument og produsentoverskudd
```

```python
def finn_og_plot_likevekt(Qd, Qs):

    # Finn likevekt
    likevekt = sp.solve(Qd - Qs, P)
    likevekt_pris = likevekt[0]
    likevekt_mengde = Qd.subs(P, likevekt_pris)
    
    # Tegn grafen
    p_verdi = np.linspace(0, 2*float(likevekt_pris), 400)
    qd_verdi = [Qd.subs(P, p_val) for p_val in p_verdi]
    qs_verdi = [Qs.subs(P, p_val) for p_val in p_verdi]
    
    plt.figure(figsize=(10,6))
    plt.plot(p_verdi, qd_verdi, label='Etterspørsel')
    plt.plot(p_verdi, qs_verdi, label='Tilbud')
    plt.axvline(x=float(likevekt_pris), color='r', linestyle='--', label=f'Likevektspris = {likevekt_pris:.2f}')
    plt.axhline(y=float(likevekt_mengde), color='g', linestyle='--', label=f'Likevektsmengde = {likevekt_mengde:.2f}')
    
    # Merking av likevektspunktet
    plt.scatter([float(likevekt_pris)], [float(likevekt_mengde)], color='purple', zorder=5)
    
    plt.xlabel('Pris')
    plt.ylabel('Mengde')
    plt.title('Tilbud og Etterspørsel for Tesla')
    plt.legend()
    plt.grid(True)
    plt.show()

    return likevekt_pris, likevekt_mengde

# Test funksjonen
likevekt_pris, likevekt_mengde = finn_og_plot_likevekt(Qd, Qs)
print(f"Likevektspris: {likevekt_pris:.2f}")
print(f"Likevektsmengde: {likevekt_mengde:.2f}")
```

```python
# Juster tilbudsfunksjonen for å reflektere avgiften
Qsmed_skatt = Qs + 50000

# Finn den nye likevekten
ny_likevekt_pris, ny_likevekt_mengde = finn_og_plot_likevekt(Qd, Qsmed_skatt)

# Beregn endringen i etterspørsel
endring_i_etterspørsel = ny_likevekt_mengde - likevekt_mengde

print(f"Opprinnelig likevektsmengde: {likevekt_mengde:.2f}")
print(f"Likevektsmengde etter pålegg av avgift: {ny_likevekt_mengde:.2f}")
print(f"Endring i etterspørsel etter pålegg av avgift: {endring_i_etterspørsel:.2f}")
```

4. Test funksjonen på minst to ulike etterspørselsfunksjoner og to ulike tilbudsfunksjoner (fire kombinasjoner totalt).

```python
def test_og_print_kombinasjon(Qd, Qs, kombinasjon_navn):
    print(f"\n{kombinasjon_navn}")
    likevekt_pris, likevekt_mengde = finn_og_plot_likevekt(Qd, Qs)
    print(f"Likevektspris for {kombinasjon_navn}: {likevekt_pris:.2f}")
    print(f"Likevektsmengde for {kombinasjon_navn}: {likevekt_mengde:.2f}")

# Ny etterspørselsfunksjon 1
Qd1 = a - b*P

# Ny etterspørselsfunksjon 2
Qd2 = a - b*sp.sqrt(P)

# Ny tilbudsfunksjon 1
Qs1 = c*P - d

# Ny tilbudsfunksjon 2
Qs2 = c*sp.sqrt(P) + d

# Tester funksjonen med de fire kombinasjonene
test_og_print_kombinasjon(Qd1, Qs1, "Kombinasjon 1: Qd1 og Qs1")
test_og_print_kombinasjon(Qd1, Qs2, "Kombinasjon 2: Qd1 og Qs2")
test_og_print_kombinasjon(Qd2, Qs1, "Kombinasjon 3: Qd2 og Qs1")
test_og_print_kombinasjon(Qd2, Qs2, "Kombinasjon 4: Qd2 og Qs2")
```

5. Lag en helt flat tilbudsfunksjon, og regn ut likevekten. Det kan argumenteres for at tilbudsfunksjonen for Tesla er flat i Norge, hva tror du argumentasjonen er for det?

```python
# Flat tilbudsfunksjon
# En flat tilbudsfunksjon antyder at uavhengig av prisen, vil antall enheter som tilbys alltid være konstant.
k = 3000
Qs_flat = sp.Integer(k)

# Finn likevekten med den flate tilbudsfunksjonen og den opprinnelige etterspørselsfunksjonen
likevekt_pris_flat, _ = finn_og_plot_likevekt(Qd, Qs_flat)
print(f"Likevektspris med flat tilbudsfunksjon: {likevekt_pris_flat}")
```
